```python
# Load the Excel file (update the filename accordingly)
file_path = "/content/Crime_Data.xlsx"
df = pd.read_excel(file_path)

# Check for duplicates based on all columns
duplicates = df[df.duplicated()]

# Save the duplicates to a new Excel file
duplicates.to_excel("duplicates.xlsx", index=False)

#print("Duplicates saved to 'duplicates.xlsx'")
print(duplicates)
```

```python
df.to_excel("cleaned_dataset.xlsx", index=False)
```

```python
# No of crimes in each year
crime_count_by_year = df.groupby('Year').size()
crime_count_by_district = df.groupby('District').size()
print(crime_count_by_year)
print(crime_count_by_district)
```

```python
#Arrest rate
arrest_rate = df['Arrest'].mean() * 100  # Percentage of arrests
print(f"Arrest Rate: {arrest_rate:.2f}%")
```

```python
#Trend analysis
import matplotlib.pyplot as plt
crime_count_by_year.plot(kind='line', title='Crime Trends by Year')
plt.show()
```

```python
#Time based plot
# Histogram for crime frequency by hour
df['Hour'] = pd.to_datetime(df['Date']).dt.hour  # Extract hour from Date
df['Hour'].plot(kind='hist', bins=24, title='Crime Frequency by Hour')
plt.xlabel('Hour of the Day')
plt.show()
```

```python
#Total crime type
crime_counts = df['Primary Type'].value_counts()

# Display the total number of crimes for each type
print("Total Crimes by Type:")
print(crime_counts)
```

```python
#Crime Type
crime_types = df['Primary Type'].value_counts()#.head(5)
crime_types.plot(kind='bar')#, title='Top 5 Crime Types')
plt.show()
```

```python
#Location pattern
import seaborn as sns
df = df.merge(crime_counts.rename('crime_counts'), left_on='Primary Type', right_index=True, how='left')

sns.boxplot(x='District', y='crime_counts', data=df)
plt.title('Crime Distribution by District')
plt.show()

```

```python
#To find out the total number of district with their names

unique_districts = df['District'].dropna().unique()  # Drop NaN values if any

# Count the total number of districts
total_districts = len(unique_districts)

# Display the total number of districts and their names
print(f"Total Number of Districts: {total_districts}")
print("District Names:")
print(unique_districts)
```
